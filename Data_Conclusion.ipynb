````python
import pandas as pd
import matplotlib.pyplot as plt
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["Spotify_Lyrics"]
collection = db["Lyrics_data"]

# Query all documents with sentiment scores and valid year
docs = list(collection.find(
    {
        "sentiment_vader": {"$exists": True},
        "sentiment_textblob": {"$exists": True},
        "year": {"$exists": True}
    },
    {
        "title": 1,
        "year": 1,
        "sentiment_vader": 1,
        "sentiment_textblob": 1
    }
))

# Check if data was found
if not docs:
    print("No documents with sentiment data found.")
else:
    # 📊 Load into DataFrame
    df = pd.DataFrame(docs)

    # 🧹 Filter for years 2019–2024
    df = df[df["year"].between(2019, 2024)]

    # ⬆️ Sort by year ascending
    df = df.sort_values(by="year", ascending=True)

    # 🧾 Print song sentiment info
    if df.empty:
        print("No data in years 2019–2024.")
    else:
        print("\n🎵 Per-song Sentiment Scores (Sorted by Year):")
        for idx, row in df.iterrows():
            print(f'{row["year"]} | {row.get("title", "Unknown Title")} | VADER: {row["sentiment_vader"]:.3f} | TextBlob: {row["sentiment_textblob"]:.3f}')

        # Plot sentiment scores
        plt.figure(figsize=(12, 6))
        plt.scatter(df["year"], df["sentiment_vader"], alpha=0.6, label="VADER Sentiment", color="blue")
        plt.scatter(df["year"], df["sentiment_textblob"], alpha=0.6, label="TextBlob Polarity", color="orange")
        plt.title("Per-song Sentiment Scores by Year (2019–2024)")
        plt.xlabel("Year")
        plt.ylabel("Sentiment Score")
        plt.legend()
        plt.grid(True)
        plt.show()

        #  print full sorted DataFrame
        print("\n Full Sorted DataFrame:")
        print(df[["year", "title", "sentiment_vader", "sentiment_textblob"]])

````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pymongo import MongoClient
from sklearn.linear_model import LinearRegression

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["Spotify_Lyrics"]
collection = db["Lyrics_data"]

# Query songs with sentiment scores and year info
docs = list(collection.find(
    {
        "sentiment_vader": {"$exists": True},
        "sentiment_textblob": {"$exists": True},
        "year": {"$exists": True}
    },
    {
        "title": 1,
        "year": 1,
        "sentiment_vader": 1,
        "sentiment_textblob": 1
    }
))

if not docs:
    print(" No documents with sentiment data found.")
else:
    # Load into DataFrame
    df = pd.DataFrame(docs)

    # Filter for 2019–2024 and sort by year
    df = df[df["year"].between(2019, 2024)]
    df = df.sort_values(by="year")

    if df.empty:
        print("No data in years 2019–2024.")
    else:
    
        # Group by year and calculate average sentiment
        yearly_sentiment = df.groupby("year")[["sentiment_vader", "sentiment_textblob"]].mean().reset_index()

        print("Yearly Average Sentiment:")
        print(yearly_sentiment)

        # Train regression model to predict 2025 sentiment (VADER)
        X = yearly_sentiment["year"].values.reshape(-1, 1)
        y = yearly_sentiment["sentiment_vader"].values
        model = LinearRegression()
        model.fit(X, y)

        # Predict sentiment for 2025
        pred_2025 = model.predict(np.array([[2025]]))
        print(f"Predicted VADER Sentiment Score for 2025: {pred_2025[0]:.4f}")

        # Plot the trend with 2025 prediction
        plt.figure(figsize=(10, 6))
        plt.plot(yearly_sentiment["year"], yearly_sentiment["sentiment_vader"], marker='o', label="Avg VADER Sentiment", color="blue")
        plt.scatter(2025, pred_2025, color='red', s=100, label="Predicted 2025")

        plt.title(" Average Yearly VADER Sentiment & 2025 Prediction")
        plt.xlabel("Year")
        plt.ylabel("Sentiment Score")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

````

## Conclusion: Music Sentiment Trends (2019-2025)

Our analysis of Spotify's most-streamed songs reveals significant insights about the relationship between cultural events and musical sentiment:

### Key Findings
1. **Pandemic Impact**:
   - 2020 showed  a low average positivity (VADER: 0.258), which coincides with COVID-19's peak, though it's not the lowest.
   - 2023 showed the lowest average positivity (VADER: 0.173). 
   - 2021 marked a strong rebound (VADER: 0.456), suggesting musical catharsis afterwards.
   - 2021, however, also displayed lower median valance (rank of positivity according to Spotify), though it gradually increased since then. 
      - This contrasts with the average positivity ranking of all the data from 2021. 
   - The median energy of each year has been gradually increasing since 2020, also suggesting a stronger positive turnaround. 
   - The number of explicit songs in each dataset has gradually increased since 2019 - primarily in the 2020-2023 years - indicating stronger feelings. 
   - Durations of songs had higher medians in 2020 and 2021 than the other years. 

2. **Current Trends**:
   - 2024 reaches peak positivity (VADER: 0.459) - the highest in our dataset, indicating more positive cultural sentiments in the year of 2024. 
   - While the number of explicit songs was high in past years, it  decreased in 2024, which might suggest a mellower overall attitude. 
   - Predicted 2025 sentiment (VADER: 0.334) suggests stabilization. 

3. **Musical Characteristics**:
   - Consistent 120 BPM tempo dominance across years
   - Extreme sentiment polarization in lyrics (e.g., 0.9966 vs -0.9936 scores)

### Interpretation
The data demonstrates music's role as a cultural barometer:
- Collective trauma (the COVID-19 years of 2020-2022) correlated with subdued sentiment in some areas (valence, energy, and explicitness), though VADER scores seemed to vary throughout those three years. 
- Recovery periods (2021-2024) showed increasing positivity in both valence and VADER score. 
- Tempo consistency suggests listeners maintain energy preferences regardless of mood.
